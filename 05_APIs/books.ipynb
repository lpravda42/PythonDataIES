```python
import pandas as pd
import requests

r = requests.get("https://www.googleapis.com/books/v1/volumes?q=fantasy&maxResults=40").json()
```

```python
df = pd.DataFrame([r["items"][i]["volumeInfo"] for i in range(len(r["items"]))])
```

```python
df.columns
```

```python
df = df[["title", "authors", "publishedDate", "categories", "averageRating"]]
df["averageRating"] = df["averageRating"].fillna(0)
```

```python
df = df.dropna(subset = ["publishedDate", "categories"])
df
```

```python
from datetime import datetime

df["publishedDate"] = pd.to_datetime(df["publishedDate"], errors = "coerce")
df
```
